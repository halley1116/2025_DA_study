## 목표변수 시각화

```python
# 프롬프트: pandas {n개의 칼럼}에 대해서 {그래프 유형}을 그릴 수 있는 코드 작성, 패키지는 matplotlib seaborn 활용
```

```python
#패키지 불러오기
import matplotlib.pyplot as plt
import seaborn as sns
# 박스 플롯 그리기
plt.figure(figsize=(8, 6)) #그래프 틀(figure) 생성
sns.boxplot(x=df["중식계"]) #seaborn에 box plot을 가져오고, 칼럼 데이터 지정
plt.title("Box Plot of Lunch") #그래프 제목
plt.ylabel("NumofLunch") #x축 제목
plt.show() #그래프 출력
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df["석식계"])
plt.title("Box Plot of Dinner")
plt.ylabel("NumofDinner")
plt.show()

#중앙값이 한 쪽에 몰려있으면 그 쪽으로 데이터가 밀집되어 있다는 의미.
```

해석 결과 정리
1. 저녁에 사람들이 많이 안 먹는 경우가 존재한다. (가설: 금요일 저녁에는 많이 안 먹지 않을까?)
2. 저녁 데이터에서 편차가 심한 분포가 보이고, 낮은 구간의 이상치가 많다. 즉, 예측이 어려울 수 있다.(*중요)

```python
# 프롬프트: pandas {1개의 칼럼}에 대해서 {dist plot}을 그릴 수 있는 코드 작성, 패키지는 matplotlib seaborn 활용
```

```python
# 예시 데이터프레임 생성

# dist plot 그리기
plt.figure(figsize=(8, 6))
sns.histplot(df['중식계'], kde=True, bins=10)
plt.title('Distribution Plot of NumofLunch')
plt.xlabel('NumofLunch')
plt.ylabel('Destiny')
plt.grid(True)
plt.show()
```

```python
# 예시 데이터프레임 생성

# dist plot 그리기
plt.figure(figsize=(8, 6))
sns.histplot(df['석식계'], kde=True, bins=10) #bin은 막대기 개수
plt.title('Distribution Plot of NumofDinner')
plt.xlabel('NumofDinner')
plt.ylabel('Destiny')
plt.grid(True) #grid는 격자
plt.show()
```

```python
df["일자"] = pd.to_datetime(df["일자"]) #pandas 공식 날짜 형식으로 변환
df = df.set_index("일자") #일자 인덱스 설정
```

```python
#라인 플롯 그리기
plt.figure(figsize=(10,6))
sns.lineplot(x=df.index, y="중식계", data=df) #"날짜" 칼럼을 x축, '값' 칼럼을 y축으로 사용
plt.title("Time Series Line Plot(NumofLunch)") #그래프 제목
plt.xlabel("Date") #X 레이블
plt.ylabel("Value") #y 레이블
plt.grid(True) # 격자 표시
plt.show()
```

```python
#라인 플롯 그리기
plt.figure(figsize=(10,6))
sns.lineplot(x=df.head(50).index, y="중식계", data=df.head(50)) #"날짜" 칼럼을 x축, '값' 칼럼을 y축으로 사용
plt.title("Time Series Line Plot(NumofLunch)") #그래프 제목
plt.xlabel("Date") #X 레이블
plt.ylabel("Value") #y 레이블
plt.grid(True) # 격자 표시
plt.show()
```

```python
#값의 빈도 계산
value_counts = df["요일"].value_counts()

#파이 차트 그리기
plt.figure(figsize=(8,6))
plt.pie(value_counts, labels=value_counts.index, startangle=140)
plt.title("Pie Chart of weekday")
plt.show()
```

```python
# 프롬프트: pandas 데이터에 대해서 상관계수 히트맵을 그릴 수 있는 코드 작성 ,패키지는 matplotlib seaborn 활용
```

```python
# 상관계수 계산
corr_matrix = df.corr(numeric_only=True)

# 히트맵 그리기
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, square=True, linewidths=0.5, linecolor='black')
plt.title('Correlation Heatmap')
plt.show()
```
